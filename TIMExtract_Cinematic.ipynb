```python
import pdfplumber
import fitz  # PyMuPDF
from PIL import Image
import io
import os

def extract_text_from_pdf(pdf_path):
    text_content = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text:
                text_content.append(text)
    return "\n".join(text_content)

def extract_images_from_pdf(pdf_path, output_folder="extracted_images"):
    os.makedirs(output_folder, exist_ok=True)
    doc = fitz.open(pdf_path)
    image_count = 0

    for page_num, page in enumerate(doc):
        images = page.get_images(full=True)

        for img_index, img in enumerate(images):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_data = base_image["image"]
            image_ext = base_image["ext"]  # Usually 'jpeg' or 'png'

            image = Image.open(io.BytesIO(image_data))
            image_path = os.path.join(output_folder, f"page_{page_num + 1}_img_{img_index + 1}.{image_ext}")
            image.save(image_path, format="JPEG")

            image_count += 1
            print(f"Saved image: {image_path}")

    print(f"Total images extracted: {image_count}")

if __name__ == "__main__":
    pdf_path = "/content/drive/MyDrive/TIM/aemr101.pdf"  # Replace with your PDF file path

    # Extract text
    extracted_text = extract_text_from_pdf(pdf_path)
    if extracted_text:
        print("Extracted Text:\n", extracted_text)
    else:
        print("No text found in the PDF.")

    # Extract images
    extract_images_from_pdf(pdf_path)
```

```python
import os
import fitz  # PyMuPDF
from PIL import Image
from io import BytesIO

def extract_text_and_images(pdf_path, output_dir):
    # Open the PDF
    pdf_document = fitz.open(pdf_path)

    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    for page_number in range(len(pdf_document)):
        page = pdf_document[page_number]
        page_folder = os.path.join(output_dir, f"page_{page_number + 1}")
        os.makedirs(page_folder, exist_ok=True)

        # --- Extract Text ---
        text = page.get_text()
        text_file_path = os.path.join(page_folder, "text.txt")
        with open(text_file_path, "w", encoding="utf-8") as text_file:
            text_file.write(text)

        # --- Extract Images ---
        image_list = page.get_images(full=True)
        for img_index, img in enumerate(image_list):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image_ext = base_image["ext"]

            # Use BytesIO to open image with PIL
            image = Image.open(BytesIO(image_bytes))
            image_path = os.path.join(page_folder, f"image_{img_index + 1}.{image_ext}")
            image.save(image_path)

        print(f"Page {page_number + 1}: extracted {len(image_list)} images and text.")

    print("\n✅ Extraction completed.")

if __name__ == "__main__":
    # Replace this with your PDF file path
    pdf_file = "/content/drive/MyDrive/TIM/aemr101.pdf"
    output_folder = "extracted_content"
    extract_text_and_images(pdf_file, output_folder)
```

```python
import os
import cv2
from PIL import Image, ImageDraw, ImageFont
from ultralytics import YOLO
import numpy as np

def load_images_and_text(page_folder):
    images = []
    texts = ""
    # Load all images
    for file in os.listdir(page_folder):
        if file.startswith("image_"):
            img_path = os.path.join(page_folder, file)
            images.append(img_path)
    # Load text
    text_file = os.path.join(page_folder, "text.txt")
    if os.path.exists(text_file):
        with open(text_file, "r", encoding="utf-8") as f:
            texts = f.read()
    return images, texts

def detect_objects_on_image(image_path, model):
    image = Image.open(image_path).convert("RGB")
    results = model.predict(source=np.array(image), save=False, verbose=False)
    return results[0]  # return prediction result

def draw_detections(image_path, results):
    image = cv2.imread(image_path)
    for box in results.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        label = results.names[int(box.cls[0])]
        cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
        cv2.putText(image, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)
    return image

def create_video(pages_dir, output_video_path, fps=1):
    # Load YOLO model
    model = YOLO("yolov8n.pt")  # Using the nano version for speed

    page_folders = sorted([os.path.join(pages_dir, folder) for folder in os.listdir(pages_dir) if folder.startswith("page_")])
    frames = []

    for page_folder in page_folders:
        images, text = load_images_and_text(page_folder)

        if not images:
            continue

        for img_path in images:
            # Detect and draw objects
            detection_result = detect_objects_on_image(img_path, model)
            img_with_detections = draw_detections(img_path, detection_result)

            # Resize for video consistency
            img_with_detections = cv2.resize(img_with_detections, (1280, 720))

            # Add text overlay
            font_scale = 0.6
            font_thickness = 1
            y0, dy = 30, 25
            lines = text.split("\n")[:10]  # Show first 10 lines max
            for i, line in enumerate(lines):
                y = y0 + i*dy
                cv2.putText(img_with_detections, line[:100], (50, y), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (0, 255, 0), font_thickness)

            frames.append(img_with_detections)

    # Save frames into video
    if frames:
        height, width, _ = frames[0].shape
        out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))
        for frame in frames:
            out.write(frame)
        out.release()
        print(f"\n✅ Video saved at: {output_video_path}")
    else:
        print("\n⚠️ No frames were generated.")

if __name__ == "__main__":
    # Example usage
    extracted_folder = "extracted_content"  # Folder created earlier
    output_video = "output_video.mp4"
    create_video(extracted_folder, output_video)
```

```python
import os
import cv2
from PIL import Image
from io import BytesIO
from gtts import gTTS
from ultralytics import YOLO
import numpy as np
from moviepy.editor import VideoFileClip, AudioFileClip, concatenate_videoclips

def load_images_and_text(page_folder):
    images = []
    texts = ""
    for file in os.listdir(page_folder):
        if file.startswith("image_"):
            img_path = os.path.join(page_folder, file)
            images.append(img_path)
    text_file = os.path.join(page_folder, "text.txt")
    if os.path.exists(text_file):
        with open(text_file, "r", encoding="utf-8") as f:
            texts = f.read()
    return images, texts

def detect_objects_on_image(image_path, model):
    image = Image.open(image_path).convert("RGB")
    results = model.predict(source=np.array(image), save=False, verbose=False)
    return results[0]

def draw_detections(image_path, results):
    image = cv2.imread(image_path)
    if results.boxes is not None:
        for box in results.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            label = results.names[int(box.cls[0])]
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(image, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)
    return image

def generate_audio(text, audio_output_path):
    tts = gTTS(text=text, lang='en')
    tts.save(audio_output_path)

def create_video_for_page(images, text, audio_path, temp_video_path, model, fps=1):
    frames = []

    for img_path in images:
        detection_result = detect_objects_on_image(img_path, model)
        img_with_detections = draw_detections(img_path, detection_result)
        img_with_detections = cv2.resize(img_with_detections, (1280, 720))

        # Overlay text
        font_scale = 0.6
        font_thickness = 1
        y0, dy = 30, 25
        lines = text.split("\n")[:10]
        for i, line in enumerate(lines):
            y = y0 + i*dy
            cv2.putText(img_with_detections, line[:100], (50, y), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (0, 255, 0), font_thickness)

        frames.append(img_with_detections)

    if frames:
        height, width, _ = frames[0].shape
        out = cv2.VideoWriter(temp_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))
        for frame in frames:
            out.write(frame)
        out.release()

    # Attach Audio
    video_clip = VideoFileClip(temp_video_path)
    audio_clip = AudioFileClip(audio_path)
    final_clip = video_clip.set_audio(audio_clip)

    output_path = temp_video_path.replace(".mp4", "_final.mp4")
    final_clip.write_videofile(output_path, codec='libx264', audio_codec='aac')
    return output_path

def create_full_video(pages_dir, output_video_path):
    model = YOLO("yolov8n.pt")
    page_folders = sorted([os.path.join(pages_dir, folder) for folder in os.listdir(pages_dir) if folder.startswith("page_")])

    clips = []

    for idx, page_folder in enumerate(page_folders):
        print(f"Processing {page_folder}...")
        images, text = load_images_and_text(page_folder)

        if not images or not text.strip():
            continue

        audio_path = os.path.join(page_folder, "page_audio.mp3")
        generate_audio(text[:1000], audio_path)  # Only take first 1000 chars for TTS (avoid huge audio)

        temp_video_path = os.path.join(page_folder, "temp_video.mp4")
        page_clip_path = create_video_for_page(images, text, audio_path, temp_video_path, model)
        clip = VideoFileClip(page_clip_path)
        clips.append(clip)

    if clips:
        final_video = concatenate_videoclips(clips, method="compose")
        final_video.write_videofile(output_video_path, codec='libx264', audio_codec='aac')
        print(f"\n✅ Full video saved at: {output_video_path}")
    else:
        print("\n⚠️ No video clips were generated.")

if __name__ == "__main__":
    # Example usage
    extracted_folder = "extracted_content"  # From previous step
    output_video = "final_output_video.mp4"
    create_full_video(extracted_folder, output_video)
```

```python
import os
import cv2
import numpy as np
from PIL import Image
from io import BytesIO
from gtts import gTTS
from moviepy.editor import (
    VideoFileClip, AudioFileClip, concatenate_videoclips,
    CompositeAudioClip, CompositeVideoClip, vfx
)
from ultralytics import YOLO
from transformers import pipeline

# Initialize once
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")  # small & fast model

def load_images_and_text(page_folder):
    images = []
    texts = ""
    for file in os.listdir(page_folder):
        if file.startswith("image_"):
            img_path = os.path.join(page_folder, file)
            images.append(img_path)
    text_file = os.path.join(page_folder, "text.txt")
    if os.path.exists(text_file):
        with open(text_file, "r", encoding="utf-8") as f:
            texts = f.read()
    return images, texts

def summarize_text(text):
    if len(text) > 100:
        summary = summarizer(text[:1000], max_length=60, min_length=25, do_sample=False)
        return summary[0]['summary_text']
    return text

def detect_objects_on_image(image_path, model):
    image = Image.open(image_path).convert("RGB")
    results = model.predict(source=np.array(image), save=False, verbose=False)
    return results[0]

def draw_detections(image_path, results):
    image = cv2.imread(image_path)
    if results.boxes is not None:
        for box in results.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            label = results.names[int(box.cls[0])]
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(image, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)
    return image

def generate_audio(text, audio_output_path):
    tts = gTTS(text=text, lang='en')
    tts.save(audio_output_path)

def create_video_for_page(images, text, audio_path, temp_video_path, model, fps=1):
    frames = []

    for img_path in images:
        detection_result = detect_objects_on_image(img_path, model)
        img_with_detections = draw_detections(img_path, detection_result)
        img_with_detections = cv2.resize(img_with_detections, (1280, 720))

        # Overlay text
        font_scale = 0.6
        font_thickness = 1
        y0, dy = 30, 25
        lines = text.split("\n")[:10]
        for i, line in enumerate(lines):
            y = y0 + i*dy
            cv2.putText(img_with_detections, line[:100], (50, y), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (0, 255, 0), font_thickness)

        frames.append(img_with_detections)

    if frames:
        height, width, _ = frames[0].shape
        out = cv2.VideoWriter(temp_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))
        for frame in frames:
            out.write(frame)
        out.release()

    # Attach Audio
    video_clip = VideoFileClip(temp_video_path)
    audio_clip = AudioFileClip(audio_path)
    final_clip = video_clip.set_audio(audio_clip)

    # Add fade-in/fade-out
    final_clip = final_clip.fx(vfx.fadein, 1).fx(vfx.fadeout, 1)

    output_path = temp_video_path.replace(".mp4", "_final.mp4")
    final_clip.write_videofile(output_path, codec='libx264', audio_codec='aac')
    return output_path

def create_full_video(pages_dir, output_video_path, background_music_path=None):
    model = YOLO("yolov8n.pt")
    page_folders = sorted([os.path.join(pages_dir, folder) for folder in os.listdir(pages_dir) if folder.startswith("page_")])

    clips = []

    for idx, page_folder in enumerate(page_folders):
        print(f"Processing {page_folder}...")
        images, text = load_images_and_text(page_folder)

        if not images or not text.strip():
            continue

        summarized_text = summarize_text(text)
        audio_path = os.path.join(page_folder, "page_audio.mp3")
        generate_audio(summarized_text, audio_path)

        temp_video_path = os.path.join(page_folder, "temp_video.mp4")
        page_clip_path = create_video_for_page(images, summarized_text, audio_path, temp_video_path, model)
        clip = VideoFileClip(page_clip_path)
        clips.append(clip)

    if clips:
        final_video = concatenate_videoclips(clips, method="compose")

        if background_music_path:
            background_music = AudioFileClip(background_music_path).volumex(0.2)  # lower volume
            final_audio = CompositeAudioClip([final_video.audio, background_music])
            final_video = final_video.set_audio(final_audio)

        final_video.write_videofile(output_video_path, codec='libx264', audio_codec='aac')
        print(f"\n✅ Full cinematic video saved at: {output_video_path}")
    else:
        print("\n⚠️ No video clips were generated.")

if __name__ == "__main__":
    extracted_folder = "extracted_content"
    output_video = "cinematic_final_output.mp4"
    background_music = "/content/drive/MyDrive/TIM/background_music.mp3"  # Path to soft background music mp3 (optional)

    create_full_video(extracted_folder, output_video, background_music_path=background_music)
```
