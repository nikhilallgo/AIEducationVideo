````python
# AI-Powered Educational Video Generator - Capstone Project
# Node-based Architecture with ONNX Runtime and QNN Support

import os
import json
import cv2
import numpy as np
import pandas as pd
from pathlib import Path
from abc import ABC, abstractmethod
from typing import Dict, List, Tuple, Any, Optional
from dataclasses import dataclass
import uuid
import logging
from datetime import datetime

# Core dependencies
import fitz  # PyMuPDF
import onnxruntime as ort
from PIL import Image, ImageDraw, ImageFont
import torch
import torchvision.transforms as transforms

# OCR and NLP - Import with error handling
import pytesseract
from transformers import pipeline, AutoModel, AutoTokenizer

# Optional OCR imports with fallback
try:
    import easyocr
    EASYOCR_AVAILABLE = False
except ImportError:
    EASYOCR_AVAILABLE = False
    print("Warning: EasyOCR not available. Install with: pip install easyocr")

try:
    import paddleocr
    PADDLEOCR_AVAILABLE = False
except ImportError:
    PADDLEOCR_AVAILABLE = False
    print("Warning: PaddleOCR not available. Install with: pip install paddleocr")

# Audio/Video processing
import librosa
import soundfile as sf
from gtts import gTTS
import moviepy.editor as mp
from pydub import AudioSegment

# Setup logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Configuration
@dataclass
class Config:
    """Configuration class for the entire pipeline"""
    pdf_path: str = "/content/drive/MyDrive/TIM/"
    output_path: str = "/content/drive/MyDrive/TIM/output/"
    temp_path: str = "/content/drive/MyDrive/TIM/temp/"

    # ONNX Runtime providers
    providers: List[str] = None

    # Model configurations
    ocr_model: str = "tesseract"  # "paddleocr", "tesseract", "easyocr"
    object_detection_model: str = "yolov8n"
    tts_model: str = "gtts"  # "gtts", "coqui", "festival"

    # Video settings
    video_fps: int = 30
    video_width: int = 1920
    video_height: int = 1080
    audio_sample_rate: int = 22050

    # Text settings
    min_text_confidence: float = 0.5
    max_words_per_frame: int = 50

    def __post_init__(self):
        if self.providers is None:
            self.providers = self._get_available_providers()

        # Create directories
        for path in [self.output_path, self.temp_path]:
            os.makedirs(path, exist_ok=True)

    def _get_available_providers(self) -> List[str]:
        """Get available ONNX Runtime providers"""
        available_providers = ort.get_available_providers()

        # Priority order: QNN (Qualcomm), CUDA, CPU
        priority_providers = [
            'QNNExecutionProvider',
            'CUDAExecutionProvider',
            'CPUExecutionProvider'
        ]

        selected_providers = []
        for provider in priority_providers:
            if provider in available_providers:
                selected_providers.append(provider)

        logger.info(f"Available providers: {available_providers}")
        logger.info(f"Selected providers: {selected_providers}")

        return selected_providers

# Base Node Class
class BaseNode(ABC):
    """Base class for all processing nodes"""

    def __init__(self, node_id: str, config: Config):
        self.node_id = node_id
        self.config = config
        self.input_data = None
        self.output_data = None
        self.metadata = {}

    @abstractmethod
    def process(self, input_data: Any) -> Any:
        """Process input data and return output"""
        pass

    def validate_input(self, input_data: Any) -> bool:
        """Validate input data"""
        return input_data is not None

    def get_metadata(self) -> Dict:
        """Get node metadata"""
        return {
            'node_id': self.node_id,
            'timestamp': datetime.now().isoformat(),
            'metadata': self.metadata
        }

# PDF Processing Node
class PDFProcessorNode(BaseNode):
    """Node for processing PDF files and extracting pages as images"""

    def __init__(self, config: Config):
        super().__init__("pdf_processor", config)

    def process(self, pdf_path: str) -> List[Dict]:
        """Extract pages from PDF as images"""
        if not os.path.exists(pdf_path):
            raise FileNotFoundError(f"PDF not found: {pdf_path}")

        doc = fitz.open(pdf_path)
        pages_data = []

        for page_num in range(len(doc)):
            page = doc.load_page(page_num)

            # Convert to image
            mat = fitz.Matrix(2, 2)  # 2x zoom
            pix = page.get_pixmap(matrix=mat)
            img_data = pix.tobytes("png")

            # Save temporary image
            temp_img_path = os.path.join(
                self.config.temp_path,
                f"page_{page_num:03d}.png"
            )

            with open(temp_img_path, "wb") as f:
                f.write(img_data)

            pages_data.append({
                'page_num': page_num,
                'image_path': temp_img_path,
                'image_data': img_data,
                'width': pix.width,
                'height': pix.height
            })

        doc.close()
        self.metadata['total_pages'] = len(pages_data)
        return pages_data

# OCR Node with multiple backends
class OCRNode(BaseNode):
    """Node for text extraction using various OCR backends"""

    def __init__(self, config: Config):
        super().__init__("ocr_processor", config)
        self.ocr_backend = self._initialize_ocr()

    def _initialize_ocr(self):
        """Initialize OCR backend based on configuration"""
        if self.config.ocr_model == "paddleocr" and PADDLEOCR_AVAILABLE:
            return paddleocr.PaddleOCR(use_angle_cls=True, lang='en')
        elif self.config.ocr_model == "easyocr" and EASYOCR_AVAILABLE:
            return easyocr.Reader(['en'])
        elif self.config.ocr_model == "tesseract":
            return "tesseract"  # Use pytesseract directly
        else:
            # Fallback to tesseract if preferred OCR is not available
            print(f"Warning: {self.config.ocr_model} not available, falling back to Tesseract")
            self.config.ocr_model = "tesseract"
            return "tesseract"

    def process(self, image_path: str) -> Dict:
        """Extract text from image using OCR"""
        if self.config.ocr_model == "paddleocr" and PADDLEOCR_AVAILABLE:
            return self._process_paddleocr(image_path)
        elif self.config.ocr_model == "easyocr" and EASYOCR_AVAILABLE:
            return self._process_easyocr(image_path)
        elif self.config.ocr_model == "tesseract":
            return self._process_tesseract(image_path)
        else:
            # Fallback to tesseract
            return self._process_tesseract(image_path)

    def _process_paddleocr(self, image_path: str) -> Dict:
        """Process with PaddleOCR"""
        if not PADDLEOCR_AVAILABLE:
            print("PaddleOCR not available, falling back to Tesseract")
            return self._process_tesseract(image_path)

        result = self.ocr_backend.ocr(image_path, cls=True)

        text_blocks = []
        full_text = ""

        for line in result[0] if result[0] else []:
            bbox = line[0]
            text = line[1][0]
            confidence = line[1][1]

            if confidence >= self.config.min_text_confidence:
                text_blocks.append({
                    'text': text,
                    'bbox': bbox,
                    'confidence': confidence,
                    'area': self._calculate_area(bbox)
                })
                full_text += text + " "

        return {
            'full_text': full_text.strip(),
            'text_blocks': text_blocks,
            'ocr_backend': 'paddleocr'
        }

    def _process_easyocr(self, image_path: str) -> Dict:
        """Process with EasyOCR"""
        if not EASYOCR_AVAILABLE:
            print("EasyOCR not available, falling back to Tesseract")
            return self._process_tesseract(image_path)

        result = self.ocr_backend.readtext(image_path)

        text_blocks = []
        full_text = ""

        for detection in result:
            bbox = detection[0]
            text = detection[1]
            confidence = detection[2]

            if confidence >= self.config.min_text_confidence:
                text_blocks.append({
                    'text': text,
                    'bbox': bbox,
                    'confidence': confidence,
                    'area': self._calculate_area(bbox)
                })
                full_text += text + " "

        return {
            'full_text': full_text.strip(),
            'text_blocks': text_blocks,
            'ocr_backend': 'easyocr'
        }

    def _process_tesseract(self, image_path: str) -> Dict:
        """Process with Tesseract OCR"""
        image = Image.open(image_path)

        # Get detailed OCR data
        ocr_data = pytesseract.image_to_data(
            image,
            output_type=pytesseract.Output.DICT
        )

        text_blocks = []
        full_text = ""

        for i in range(len(ocr_data['text'])):
            text = ocr_data['text'][i].strip()
            confidence = int(ocr_data['conf'][i])

            if text and confidence >= (self.config.min_text_confidence * 100):
                x, y, w, h = (
                    ocr_data['left'][i],
                    ocr_data['top'][i],
                    ocr_data['width'][i],
                    ocr_data['height'][i]
                )

                bbox = [[x, y], [x + w, y], [x + w, y + h], [x, y + h]]

                text_blocks.append({
                    'text': text,
                    'bbox': bbox,
                    'confidence': confidence / 100.0,
                    'area': w * h
                })
                full_text += text + " "

        return {
            'full_text': full_text.strip(),
            'text_blocks': text_blocks,
            'ocr_backend': 'tesseract'
        }

    def _calculate_area(self, bbox: List) -> float:
        """Calculate area of bounding box"""
        if len(bbox) == 4:
            x_coords = [point[0] for point in bbox]
            y_coords = [point[1] for point in bbox]
            width = max(x_coords) - min(x_coords)
            height = max(y_coords) - min(y_coords)
            return width * height
        return 0

# Object Detection Node with ONNX Runtime
class ObjectDetectionNode(BaseNode):
    """Node for object detection using ONNX Runtime"""

    def __init__(self, config: Config):
        super().__init__("object_detector", config)
        self.model = self._load_model()
        self.class_names = self._get_class_names()

    def _load_model(self):
        """Load ONNX model with appropriate providers"""
        model_path = self._download_model()

        # Create ONNX Runtime session
        session_options = ort.SessionOptions()
        session_options.graph_optimization_level = ort.GraphOptimizationLevel.ORT_ENABLE_ALL

        providers = self.config.providers
        if 'QNNExecutionProvider' in providers:
            # QNN provider options for Qualcomm NPU
            qnn_options = {
                'backend_path': '/usr/lib/libQnnHtp.so',  # Adjust path as needed
                'profiling_level': '0'
            }
            providers = [('QNNExecutionProvider', qnn_options)] + providers[1:]

        session = ort.InferenceSession(
            model_path,
            sess_options=session_options,
            providers=providers
        )

        logger.info(f"Model loaded with providers: {session.get_providers()}")
        return session

    def _download_model(self) -> str:
        """Download or locate the ONNX model"""
        model_path = os.path.join(self.config.temp_path, "yolov8n.onnx")

        if not os.path.exists(model_path):
            logger.info("Downloading YOLOv8n ONNX model...")
            # In practice, you would download from a model repository
            # For now, assume model is available or convert from PyTorch
            try:
                from ultralytics import YOLO
                model = YOLO('yolov8n.pt')
                model.export(format='onnx', dynamic=True)
                # Move to temp directory
                os.rename('yolov8n.onnx', model_path)
            except Exception as e:
                logger.error(f"Failed to download model: {e}")
                raise

        return model_path

    def _get_class_names(self) -> List[str]:
        """Get COCO class names"""
        return [
            'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus', 'train',
            'truck', 'boat', 'traffic light', 'fire hydrant', 'stop sign',
            'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep',
            'cow', 'elephant', 'bear', 'zebra', 'giraffe', 'backpack', 'umbrella',
            'handbag', 'tie', 'suitcase', 'frisbee', 'skis', 'snowboard',
            'sports ball', 'kite', 'baseball bat', 'baseball glove', 'skateboard',
            'surfboard', 'tennis racket', 'bottle', 'wine glass', 'cup', 'fork',
            'knife', 'spoon', 'bowl', 'banana', 'apple', 'sandwich', 'orange',
            'broccoli', 'carrot', 'hot dog', 'pizza', 'donut', 'cake', 'chair',
            'couch', 'potted plant', 'bed', 'dining table', 'toilet', 'tv',
            'laptop', 'mouse', 'remote', 'keyboard', 'cell phone', 'microwave',
            'oven', 'toaster', 'sink', 'refrigerator', 'book', 'clock', 'vase',
            'scissors', 'teddy bear', 'hair drier', 'toothbrush'
        ]

    def process(self, image_path: str) -> Dict:
        """Detect objects in image"""
        image = cv2.imread(image_path)
        if image is None:
            raise ValueError(f"Could not load image: {image_path}")

        # Preprocess image
        input_tensor = self._preprocess_image(image)

        # Run inference
        input_name = self.model.get_inputs()[0].name
        outputs = self.model.run(None, {input_name: input_tensor})

        # Post-process results
        detections = self._postprocess_outputs(outputs[0], image.shape)

        return {
            'detections': detections,
            'image_shape': image.shape,
            'detection_count': len(detections)
        }

    def _preprocess_image(self, image: np.ndarray) -> np.ndarray:
        """Preprocess image for ONNX model"""
        # Resize to model input size (640x640 for YOLOv8)
        resized = cv2.resize(image, (640, 640))

        # Convert BGR to RGB
        rgb_image = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)

        # Normalize and transpose
        normalized = rgb_image.astype(np.float32) / 255.0
        transposed = np.transpose(normalized, (2, 0, 1))

        # Add batch dimension
        input_tensor = np.expand_dims(transposed, axis=0)

        return input_tensor

    def _postprocess_outputs(self, outputs: np.ndarray, original_shape: Tuple) -> List[Dict]:
        """Post-process model outputs"""
        detections = []

        # YOLOv8 output format: [batch, 84, 8400] where 84 = 4 + 80 classes
        outputs = outputs.squeeze()  # Remove batch dimension
        outputs = outputs.T  # Transpose to [8400, 84]

        # Extract boxes and scores
        boxes = outputs[:, :4]  # x, y, w, h
        scores = outputs[:, 4:]  # class scores

        # Get class predictions
        class_ids = np.argmax(scores, axis=1)
        confidences = np.max(scores, axis=1)

        # Filter by confidence threshold
        valid_indices = confidences > 0.5

        boxes = boxes[valid_indices]
        class_ids = class_ids[valid_indices]
        confidences = confidences[valid_indices]

        # Scale boxes to original image size
        h, w = original_shape[:2]
        scale_x, scale_y = w / 640, h / 640

        for i in range(len(boxes)):
            x, y, width, height = boxes[i]

            # Convert center coordinates to top-left
            x1 = int((x - width / 2) * scale_x)
            y1 = int((y - height / 2) * scale_y)
            x2 = int((x + width / 2) * scale_x)
            y2 = int((y + height / 2) * scale_y)

            detection = {
                'class_id': int(class_ids[i]),
                'class_name': self.class_names[class_ids[i]],
                'confidence': float(confidences[i]),
                'bbox': [x1, y1, x2, y2],
                'area': (x2 - x1) * (y2 - y1)
            }
            detections.append(detection)

        return detections

# Text-to-Speech Node
class TTSNode(BaseNode):
    """Node for converting text to speech"""

    def __init__(self, config: Config):
        super().__init__("tts_processor", config)

    def process(self, text: str, output_path: str) -> Dict:
        """Convert text to speech"""
        if not text.strip():
            return {'audio_path': None, 'duration': 0}

        # Generate speech
        tts = gTTS(text=text, lang='en', slow=False)
        temp_audio_path = os.path.join(
            self.config.temp_path,
            f"tts_{uuid.uuid4().hex}.mp3"
        )

        tts.save(temp_audio_path)

        # Convert to WAV and get duration
        audio = AudioSegment.from_mp3(temp_audio_path)
        audio.export(output_path, format="wav")

        # Clean up temp file
        os.remove(temp_audio_path)

        return {
            'audio_path': output_path,
            'duration': len(audio) / 1000.0  # Duration in seconds
        }

# Content Matching Node
class ContentMatchingNode(BaseNode):
    """Node for matching text with detected objects"""

    def __init__(self, config: Config):
        super().__init__("content_matcher", config)

    def process(self, ocr_data: Dict, detection_data: Dict) -> Dict:
        """Match text content with detected objects"""
        text_blocks = ocr_data.get('text_blocks', [])
        detections = detection_data.get('detections', [])

        matches = []
        unmatched_text = []
        unmatched_objects = detections.copy()

        for text_block in text_blocks:
            text = text_block['text'].lower()
            text_bbox = text_block['bbox']

            best_match = None
            best_score = 0

            for detection in detections:
                object_name = detection['class_name'].lower()
                object_bbox = detection['bbox']

                # Calculate semantic similarity
                semantic_score = self._calculate_semantic_similarity(text, object_name)

                # Calculate spatial proximity
                spatial_score = self._calculate_spatial_proximity(text_bbox, object_bbox)

                # Combined score
                combined_score = (semantic_score * 0.7) + (spatial_score * 0.3)

                if combined_score > best_score and combined_score > 0.3:
                    best_match = detection
                    best_score = combined_score

            if best_match:
                matches.append({
                    'text': text_block['text'],
                    'text_bbox': text_bbox,
                    'object': best_match,
                    'confidence': best_score
                })
                if best_match in unmatched_objects:
                    unmatched_objects.remove(best_match)
            else:
                unmatched_text.append(text_block)

        return {
            'matches': matches,
            'unmatched_text': unmatched_text,
            'unmatched_objects': unmatched_objects,
            'match_count': len(matches)
        }

    def _calculate_semantic_similarity(self, text: str, object_name: str) -> float:
        """Calculate semantic similarity between text and object name"""
        # Simple keyword matching (can be enhanced with embeddings)
        text_words = set(text.lower().split())
        object_words = set(object_name.lower().split())

        # Direct match
        if object_name in text or any(word in text for word in object_words):
            return 1.0

        # Partial match
        common_words = text_words.intersection(object_words)
        if common_words:
            return len(common_words) / max(len(text_words), len(object_words))

        return 0.0

    def _calculate_spatial_proximity(self, text_bbox: List, object_bbox: List) -> float:
        """Calculate spatial proximity between text and object"""
        # Convert bounding boxes to center points
        text_center = self._get_bbox_center(text_bbox)
        object_center = self._get_bbox_center(object_bbox)

        # Calculate Euclidean distance
        distance = np.sqrt(
            (text_center[0] - object_center[0]) ** 2 +
            (text_center[1] - object_center[1]) ** 2
        )

        # Normalize distance (assuming image dimensions)
        max_distance = np.sqrt(1920**2 + 1080**2)  # Diagonal of typical image
        normalized_distance = distance / max_distance

        # Convert to proximity score (closer = higher score)
        proximity_score = 1 - normalized_distance

        return max(0, proximity_score)

    def _get_bbox_center(self, bbox: List) -> Tuple[float, float]:
        """Get center point of bounding box"""
        if len(bbox) == 4 and isinstance(bbox[0], list):
            # OCR format: [[x1, y1], [x2, y2], [x3, y3], [x4, y4]]
            x_coords = [point[0] for point in bbox]
            y_coords = [point[1] for point in bbox]
            center_x = sum(x_coords) / len(x_coords)
            center_y = sum(y_coords) / len(y_coords)
        else:
            # Detection format: [x1, y1, x2, y2]
            center_x = (bbox[0] + bbox[2]) / 2
            center_y = (bbox[1] + bbox[3]) / 2

        return (center_x, center_y)

# Video Generation Node
class VideoGenerationNode(BaseNode):
    """Node for generating videos from processed content"""

    def __init__(self, config: Config):
        super().__init__("video_generator", config)

    def process(self, pages_data: List[Dict], output_path: str) -> Dict:
        """Generate video from processed pages"""
        video_clips = []

        for page_data in pages_data:
            # Create video clip for this page
            clip = self._create_page_clip(page_data)
            if clip:
                video_clips.append(clip)

        if not video_clips:
            raise ValueError("No valid video clips generated")

        # Concatenate all clips
        final_video = mp.concatenate_videoclips(video_clips)

        # Write final video
        final_video.write_videofile(
            output_path,
            fps=self.config.video_fps,
            codec='libx264',
            audio_codec='aac'
        )

        # Clean up
        for clip in video_clips:
            clip.close()
        final_video.close()

        return {
            'video_path': output_path,
            'duration': sum(clip.duration for clip in video_clips),
            'total_pages': len(pages_data)
        }

    def _create_page_clip(self, page_data: Dict) -> Optional[mp.VideoClip]:
        """Create video clip for a single page"""
        try:
            # Load page image
            image_clip = mp.ImageClip(page_data['image_path'])

            # Load audio if available
            audio_path = page_data.get('audio_path')
            if audio_path and os.path.exists(audio_path):
                audio_clip = mp.AudioFileClip(audio_path)

                # Set video duration to match audio
                image_clip = image_clip.set_duration(audio_clip.duration)

                # Add audio to video
                video_clip = image_clip.set_audio(audio_clip)
            else:
                # Default duration if no audio
                video_clip = image_clip.set_duration(3.0)

            # Resize to target resolution
            video_clip = video_clip.resize(
                (self.config.video_width, self.config.video_height)
            )

            return video_clip

        except Exception as e:
            logger.error(f"Error creating clip for page {page_data.get('page_num')}: {e}")
            return None

# Pipeline Controller
class VideoPipeline:
    """Main pipeline controller for the video generation process"""

    def __init__(self, config: Config):
        self.config = config
        self.nodes = self._initialize_nodes()
        self.pipeline_data = {}

    def _initialize_nodes(self) -> Dict[str, BaseNode]:
        """Initialize all processing nodes"""
        return {
            'pdf_processor': PDFProcessorNode(self.config),
            'ocr_processor': OCRNode(self.config),
            'object_detector': ObjectDetectionNode(self.config),
            'content_matcher': ContentMatchingNode(self.config),
            'tts_processor': TTSNode(self.config),
            'video_generator': VideoGenerationNode(self.config)
        }

    def process_pdf(self, pdf_filename: str) -> str:
        """Process a PDF file and generate educational video"""
        pdf_path = os.path.join(self.config.pdf_path, pdf_filename)
        output_video_path = os.path.join(
            self.config.output_path,
            f"{Path(pdf_filename).stem}_educational_video.mp4"
        )

        logger.info(f"Processing PDF: {pdf_filename}")

        try:
            # Step 1: Extract pages from PDF
            logger.info("Extracting pages from PDF...")
            pages_data = self.nodes['pdf_processor'].process(pdf_path)

            # Step 2: Process each page
            processed_pages = []
            for page_data in pages_data:
                logger.info(f"Processing page {page_data['page_num']}...")

                # OCR processing
                ocr_result = self.nodes['ocr_processor'].process(page_data['image_path'])

                # Object detection
                detection_result = self.nodes['object_detector'].process(page_data['image_path'])

                # Content matching
                matching_result = self.nodes['content_matcher'].process(
                    ocr_result, detection_result
                )

                # Generate explanatory text
                explanation_text = self._generate_explanation_text(
                    ocr_result, detection_result, matching_result
                )

                # Text-to-speech
                audio_path = os.path.join(
                    self.config.temp_path,
                    f"page_{page_data['page_num']}_audio.wav"
                )

                tts_result = self.nodes['tts_processor'].process(
                    explanation_text, audio_path
                )

                # Combine all data for this page
                page_data.update({
                    'ocr_result': ocr_result,
                    'detection_result': detection_result,
                    'matching_result': matching_result,
                    'explanation_text': explanation_text,
                    'audio_path': tts_result['audio_path'],
                    'audio_duration': tts_result['duration']
                })

                processed_pages.append(page_data)

            # Step 3: Generate final video
            logger.info("Generating final video...")
            video_result = self.nodes['video_generator'].process(
                processed_pages, output_video_path
            )

            logger.info(f"Video generated successfully: {output_video_path}")

            return output_video_path

        except Exception as e:
            logger.error(f"Error processing PDF: {e}")
            raise

    def _generate_explanation_text(self, ocr_result: Dict, detection_result: Dict,
                                 matching_result: Dict) -> str:
        """Generate educational explanation text"""
        explanation_parts = []

        # Start with general page description
        explanation_parts.append("Let's explore this page together!")

        # Describe detected objects
        detections = detection_result.get('detections', [])
        if detections:
            object_names = [det['class_name'] for det in detections]
            unique_objects = list(set(object_names))

            if len(unique_objects) == 1:
                explanation_parts.append(f"I can see a {unique_objects[0]} in this image.")
            elif len(unique_objects) == 2:
                explanation_parts.append(f"I can see a {unique_objects[0]} and a {unique_objects[1]} in this image.")
            else:
                explanation_parts.append(f"I can see several things including {', '.join(unique_objects[:3])} in this image.")

        # Describe text content
        full_text = ocr_result.get('full_text', '').strip()
        if full_text:
            # Simplify text for Grade 1 kids
            simplified_text = self._simplify_text_for_kids(full_text)
            explanation_parts.append(f"The text on this page says: {simplified_text}")

        # Describe matches between text and objects
        matches = matching_result.get('matches', [])
        if matches:
            for match in matches[:3]:  # Limit to top 3 matches
                text = match['text']
                object_name = match['object']['class_name']
                explanation_parts.append(f"The word '{text}' is related to the {object_name} we can see.")

        # Add educational encouragement
        explanation_parts.append("Great job looking at this page! Let's continue learning together.")

        return " ".join(explanation_parts)

    def _simplify_text_for_kids(self, text: str) -> str:
        """Simplify text content for Grade 1 children"""
        # Limit to reasonable length for young kids
        words = text.split()
        if len(words) > self.config.max_words_per_frame:
            words = words[:self.config.max_words_per_frame]

        # Join back and clean up
        simplified = " ".join(words)
        simplified = simplified.replace('\n', ' ').strip()

        return simplified

    def get_pipeline_status(self) -> Dict:
        """Get current pipeline status"""
        return {
            'config': self.config.__dict__,
            'nodes': list(self.nodes.keys()),
            'pipeline_data': self.pipeline_data
        }

# Utility Functions
def setup_environment():
    """Setup the environment and check dependencies"""
    print("Setting up AI Educational Video Generator...")

    # Check ONNX Runtime providers
    available_providers = ort.get_available_providers()
    print(f"Available ONNX Runtime providers: {available_providers}")

    # Check if QNN provider is available
    if 'QNNExecutionProvider' in available_providers:
        print("✓ Qualcomm Neural Network (QNN) provider is available")
    else:
        print("! QNN provider not available, falling back to other providers")

    # Check GPU availability
    if 'CUDAExecutionProvider' in available_providers:
        print("✓ CUDA GPU provider is available")
    else:
        print("! CUDA provider not available")

    print("Environment setup complete!")

# Example usage and main execution
def main():
    """Main execution function"""
    # Setup environment
    setup_environment()

    # Create configuration with fallback OCR
    config = Config(
        pdf_path="/content/drive/MyDrive/TIM/",
        output_path="/content/drive/MyDrive/TIM/output/",
        temp_path="/content/drive/MyDrive/TIM/temp/",
        ocr_model="tesseract",  # Default to tesseract as it's most reliable
        video_fps=24,
        video_width=1280,
        video_height=720
    )

    # Auto-detect best available OCR
    if PADDLEOCR_AVAILABLE:
        config.ocr_model = "paddleocr"
        print("Using PaddleOCR for text extraction")
    elif EASYOCR_AVAILABLE:
        config.ocr_model = "easyocr"
        print("Using EasyOCR for text extraction")
    else:
        config.ocr_model = "tesseract"
        print("Using Tesseract for text extraction")

    # Initialize pipeline
    pipeline = VideoPipeline(config)

    # Process PDF files
    pdf_files = [f for f in os.listdir(config.pdf_path) if f.endswith('.pdf')]

    if not pdf_files:
        print("No PDF files found in the specified directory!")
        return

    print(f"Found {len(pdf_files)} PDF files to process")

    for pdf_file in pdf_files:
        try:
            print(f"\nProcessing: {pdf_file}")
            output_video = pipeline.process_pdf(pdf_file)
            print(f"✓ Successfully generated video: {output_video}")
        except Exception as e:
            print(f"✗ Failed to process {pdf_file}: {e}")

    print("\nAll processing complete!")

# Advanced Features and Extensions
class ModelManager:
    """Manager for handling model updates and replacements"""

    def __init__(self, config: Config):
        self.config = config
        self.model_registry = {}
        self.model_versions = {}

    def register_model(self, model_name: str, model_path: str, version: str):
        """Register a new model version"""
        if model_name not in self.model_registry:
            self.model_registry[model_name] = {}

        self.model_registry[model_name][version] = model_path
        self.model_versions[model_name] = version

    def get_model_path(self, model_name: str, version: str = None) -> str:
        """Get model path for specified version"""
        if model_name not in self.model_registry:
            raise ValueError(f"Model {model_name} not registered")

        if version is None:
            version = self.model_versions.get(model_name)

        if version not in self.model_registry[model_name]:
            raise ValueError(f"Version {version} not found for model {model_name}")

        return self.model_registry[model_name][version]

    def update_model(self, model_name: str, new_version: str):
        """Update to a new model version"""
        if model_name in self.model_registry and new_version in self.model_registry[model_name]:
            self.model_versions[model_name] = new_version
            print(f"Updated {model_name} to version {new_version}")
        else:
            raise ValueError(f"Model {model_name} version {new_version} not found")

class PerformanceMonitor:
    """Monitor pipeline performance and resource usage"""

    def __init__(self):
        self.metrics = {}
        self.start_times = {}

    def start_timer(self, operation: str):
        """Start timing an operation"""
        self.start_times[operation] = datetime.now()

    def end_timer(self, operation: str):
        """End timing and record duration"""
        if operation in self.start_times:
            duration = (datetime.now() - self.start_times[operation]).total_seconds()
            if operation not in self.metrics:
                self.metrics[operation] = []
            self.metrics[operation].append(duration)
            del self.start_times[operation]
            return duration
        return 0

    def get_average_time(self, operation: str) -> float:
        """Get average time for an operation"""
        if operation in self.metrics:
            return sum(self.metrics[operation]) / len(self.metrics[operation])
        return 0

    def get_performance_report(self) -> Dict:
        """Get performance report"""
        report = {}
        for operation, times in self.metrics.items():
            report[operation] = {
                'count': len(times),
                'total_time': sum(times),
                'average_time': sum(times) / len(times),
                'min_time': min(times),
                'max_time': max(times)
            }
        return report

# Batch Processing Support
class BatchProcessor:
    """Handle batch processing of multiple PDFs"""

    def __init__(self, config: Config):
        self.config = config
        self.pipeline = VideoPipeline(config)
        self.performance_monitor = PerformanceMonitor()

    def process_batch(self, pdf_directory: str, output_directory: str = None) -> List[Dict]:
        """Process multiple PDFs in batch"""
        if output_directory is None:
            output_directory = self.config.output_path

        pdf_files = [f for f in os.listdir(pdf_directory) if f.endswith('.pdf')]
        results = []

        for i, pdf_file in enumerate(pdf_files):
            print(f"Processing batch {i+1}/{len(pdf_files)}: {pdf_file}")

            try:
                self.performance_monitor.start_timer(f"pdf_processing_{pdf_file}")

                # Update config for this PDF
                self.config.pdf_path = pdf_directory
                self.config.output_path = output_directory

                # Process PDF
                output_video = self.pipeline.process_pdf(pdf_file)

                duration = self.performance_monitor.end_timer(f"pdf_processing_{pdf_file}")

                results.append({
                    'pdf_file': pdf_file,
                    'output_video': output_video,
                    'status': 'success',
                    'processing_time': duration
                })

            except Exception as e:
                results.append({
                    'pdf_file': pdf_file,
                    'status': 'error',
                    'error': str(e)
                })

        return results

# Configuration for different deployment scenarios
class DeploymentConfig:
    """Configuration for different deployment scenarios"""

    @staticmethod
    def get_cloud_config() -> Config:
        """Configuration for cloud deployment"""
        return Config(
            pdf_path="/app/data/input/",
            output_path="/app/data/output/",
            temp_path="/app/data/temp/",
            providers=['CUDAExecutionProvider', 'CPUExecutionProvider'],
            video_fps=30,
            video_width=1920,
            video_height=1080
        )

    @staticmethod
    def get_edge_config() -> Config:
        """Configuration for edge device deployment"""
        return Config(
            pdf_path="/data/input/",
            output_path="/data/output/",
            temp_path="/data/temp/",
            providers=['QNNExecutionProvider', 'CPUExecutionProvider'],
            video_fps=24,
            video_width=1280,
            video_height=720,
            ocr_model="paddleocr"  # More efficient for edge devices
        )

    @staticmethod
    def get_development_config() -> Config:
        """Configuration for development environment"""
        return Config(
            pdf_path="./data/input/",
            output_path="./data/output/",
            temp_path="./data/temp/",
            providers=['CPUExecutionProvider'],
            video_fps=24,
            video_width=1280,
            video_height=720
        )

# Quality Assessment Node
class QualityAssessmentNode(BaseNode):
    """Node for assessing the quality of generated content"""

    def __init__(self, config: Config):
        super().__init__("quality_assessor", config)

    def process(self, processed_data: Dict) -> Dict:
        """Assess quality of processed content"""
        quality_scores = {}

        # OCR quality assessment
        ocr_result = processed_data.get('ocr_result', {})
        quality_scores['ocr_quality'] = self._assess_ocr_quality(ocr_result)

        # Object detection quality
        detection_result = processed_data.get('detection_result', {})
        quality_scores['detection_quality'] = self._assess_detection_quality(detection_result)

        # Content matching quality
        matching_result = processed_data.get('matching_result', {})
        quality_scores['matching_quality'] = self._assess_matching_quality(matching_result)

        # Overall quality score
        quality_scores['overall_quality'] = np.mean(list(quality_scores.values()))

        return quality_scores

    def _assess_ocr_quality(self, ocr_result: Dict) -> float:
        """Assess OCR quality based on confidence and text coherence"""
        text_blocks = ocr_result.get('text_blocks', [])

        if not text_blocks:
            return 0.0

        # Average confidence score
        confidences = [block['confidence'] for block in text_blocks]
        avg_confidence = np.mean(confidences)

        # Text coherence (simple heuristic)
        full_text = ocr_result.get('full_text', '')
        coherence_score = min(1.0, len(full_text.split()) / 10)  # Normalize by word count

        return (avg_confidence * 0.7) + (coherence_score * 0.3)

    def _assess_detection_quality(self, detection_result: Dict) -> float:
        """Assess object detection quality"""
        detections = detection_result.get('detections', [])

        if not detections:
            return 0.0

        # Average confidence score
        confidences = [det['confidence'] for det in detections]
        avg_confidence = np.mean(confidences)

        # Detection count score (more objects generally better for educational content)
        count_score = min(1.0, len(detections) / 5)  # Normalize by expected object count

        return (avg_confidence * 0.8) + (count_score * 0.2)

    def _assess_matching_quality(self, matching_result: Dict) -> float:
        """Assess content matching quality"""
        matches = matching_result.get('matches', [])
        total_text = len(matching_result.get('unmatched_text', []))
        total_objects = len(matching_result.get('unmatched_objects', []))

        if not matches:
            return 0.0

        # Match confidence
        match_confidences = [match['confidence'] for match in matches]
        avg_match_confidence = np.mean(match_confidences)

        # Match ratio (how many items were successfully matched)
        total_items = len(matches) + total_text + total_objects
        match_ratio = len(matches) / max(1, total_items)

        return (avg_match_confidence * 0.6) + (match_ratio * 0.4)

# Installation and Setup Instructions
INSTALLATION_INSTRUCTIONS = """
# AI Educational Video Generator - Installation Instructions

## Prerequisites

### 1. Python Environment
```bash
# Create virtual environment
python -m venv venv
source venv/bin/activate  # On Windows: venv\\Scripts\\activate

# Install Python dependencies
pip install -r requirements.txt
```

### 2. Requirements.txt
```
# Core dependencies
numpy>=1.21.0
opencv-python>=4.5.0
Pillow>=8.0.0
torch>=1.11.0
torchvision>=0.12.0
onnxruntime>=1.12.0
onnxruntime-gpu>=1.12.0  # For GPU support

# OCR dependencies (install as needed)
pytesseract>=0.3.9  # Required - most reliable OCR
paddleocr>=2.6.0    # Optional - better accuracy
easyocr>=1.6.0      # Optional - good for multilingual

# NLP dependencies
transformers>=4.20.0
torch-audio>=0.11.0

# Audio/Video processing
moviepy>=1.0.3
gtts>=2.2.4
pydub>=0.25.1
librosa>=0.9.0
soundfile>=0.10.0

# PDF processing
PyMuPDF>=1.20.0

# Utilities
pandas>=1.4.0
requests>=2.28.0
tqdm>=4.64.0
ultralytics>=8.0.0  # For YOLOv8 model export
```

### 3. OCR Backend Installation

#### Option 1: Tesseract (Recommended - Most Reliable)
```bash
# Ubuntu/Debian
sudo apt-get install tesseract-ocr tesseract-ocr-eng

# macOS
brew install tesseract

# Windows
# Download from: https://github.com/UB-Mannheim/tesseract/wiki
```

#### Option 2: PaddleOCR (Better Accuracy)
```bash
# Install PaddlePaddle first
pip install paddlepaddle-gpu  # For GPU
# OR
pip install paddlepaddle      # For CPU only

# Then install PaddleOCR
pip install paddleocr
```

#### Option 3: EasyOCR (Good for Multilingual)
```bash
pip install easyocr
```

### 3. System Dependencies

#### Ubuntu/Debian:
```bash
# Install system dependencies
sudo apt-get update
sudo apt-get install -y \\
    tesseract-ocr \\
    tesseract-ocr-eng \\
    ffmpeg \\
    libsm6 \\
    libxext6 \\
    libfontconfig1 \\
    libxrender1 \\
    libgl1-mesa-glx
```

#### For Qualcomm QNN Support:
```bash
# Install Qualcomm Neural Network SDK
# Download from Qualcomm Developer Network
# Follow platform-specific installation guide
```

## Usage Examples

### Basic Usage:
```python
from video_generator import Config, VideoPipeline

# Initialize configuration
config = Config(
    pdf_path="/path/to/pdfs/",
    output_path="/path/to/output/",
    ocr_model="paddleocr"
)

# Create pipeline
pipeline = VideoPipeline(config)

# Process single PDF
output_video = pipeline.process_pdf("sample.pdf")
print(f"Generated video: {output_video}")
```

### Batch Processing:
```python
from video_generator import BatchProcessor, Config

config = Config()
batch_processor = BatchProcessor(config)

# Process all PDFs in directory
results = batch_processor.process_batch("/path/to/pdfs/")

for result in results:
    print(f"PDF: {result['pdf_file']}, Status: {result['status']}")
```

### Model Management:
```python
from video_generator import ModelManager, Config

config = Config()
model_manager = ModelManager(config)

# Register new model version
model_manager.register_model(
    "yolov8",
    "/path/to/yolov8n_v2.onnx",
    "v2.0"
)

# Update to new version
model_manager.update_model("yolov8", "v2.0")
```

## Configuration Options

### Hardware-Specific Configurations:
```python
# For Qualcomm NPU devices
config = Config(
    providers=['QNNExecutionProvider', 'CPUExecutionProvider']
)

# For NVIDIA GPU devices
config = Config(
    providers=['CUDAExecutionProvider', 'CPUExecutionProvider']
)

# For CPU-only devices
config = Config(
    providers=['CPUExecutionProvider']
)
```

### Quality vs Performance Trade-offs:
```python
# High quality (slower processing)
config = Config(
    video_fps=30,
    video_width=1920,
    video_height=1080,
    min_text_confidence=0.8
)

# Fast processing (lower quality)
config = Config(
    video_fps=24,
    video_width=1280,
    video_height=720,
    min_text_confidence=0.5
)
```

## Troubleshooting

### Common Issues:

1. **ONNX Runtime Provider Issues:**
   - Ensure correct ONNX Runtime version for your hardware
   - Check provider availability with `ort.get_available_providers()`

2. **Memory Issues:**
   - Reduce video resolution
   - Process PDFs in smaller batches
   - Enable GPU memory optimization

3. **OCR Accuracy Issues:**
   - Try different OCR backends (PaddleOCR, EasyOCR, Tesseract)
   - Preprocess images for better OCR results

4. **Audio Synchronization:**
   - Check audio file generation
   - Verify MoviePy audio codec settings

## Performance Optimization

### For Edge Devices:
- Use quantized ONNX models
- Enable QNN provider for Qualcomm devices
- Reduce video resolution and frame rate

### For Cloud Deployment:
- Use GPU acceleration
- Enable batch processing
- Implement caching for repeated operations

### Memory Management:
- Clean up temporary files regularly
- Use context managers for resource handling
- Monitor memory usage during processing
"""

if __name__ == "__main__":
    print("AI Educational Video Generator")
    print("=" * 50)

    # Print installation instructions
    print(INSTALLATION_INSTRUCTIONS)

    # Run main pipeline
    main()
````
